```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_cols = train.select_dtypes(include=['object']).columns
for col in categorical_cols:
    train[col] = le.fit_transform(train[[col]])
    test[col] = le.transform(test[[col]])
train    
```

```python
sns.heatmap(train.corr(),annot=True,fmt=".1f")
```

```python
Y=train['accident_risk']
train.drop('accident_risk',axis=1,inplace=True)
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor ,GradientBoostingRegressor,HistGradientBoostingRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
X_train,X_valid,Y_train,Y_test=train_test_split(train[['curvature','speed_limit','lighting','weather','num_reported_accidents','holiday']],Y,train_size=0.85,test_size=0.15,random_state=42,shuffle=True)
model0=LGBMRegressor(n_estimators=500,random_state=42)
model1=HistGradientBoostingRegressor(max_iter=1000, learning_rate=0.05,random_state=42)
model2=RandomForestRegressor(n_estimators=500,random_state=42)
#model2=DecisionTreeRegressor(max_depth=50,random_state=42,splitter='best')
model3=XGBRegressor(n_estimators=500,random_state=42)
model4=CatBoostRegressor(n_estimators=500,random_state=42,learning_rate=0.05)
```

```python
model0.fit(X_train,Y_train)
```

```python
model1.fit(X_train,Y_train)
```

```python
model2.fit(train[['curvature','speed_limit','lighting','weather','num_reported_accidents','holiday']],Y)
```

```python
model3.fit(X_train,Y_train)
```

```python
model4.fit(X_train,Y_train)
```

```python
pred0=model0.predict(X_valid)
print(pred0)
print(root_mean_squared_error(Y_test,pred0))
```

```python
pred1=model1.predict(X_valid)
print(pred1)
print(root_mean_squared_error(Y_test,pred1))
```

```python
pred2=model2.predict(X_valid)
print(pred2)
print(root_mean_squared_error(Y_test,pred2))
```

```python
pred3=model3.predict(X_valid)
print(pred3)
print(root_mean_squared_error(Y_test,pred3))
```

```python
pred4=model4.predict(X_valid)
print(pred4)
print(root_mean_squared_error(Y_test,pred4))
```
